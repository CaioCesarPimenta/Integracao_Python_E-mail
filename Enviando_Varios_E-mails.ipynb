```python
import pandas as pd
import  win32com.client as win32
outlook = win32.Dispatch('outlook.application')
gerentes = pd.read_excel('Enviar E-mails.xlsx')
display(gerentes)
```

```python



for i, loop in enumerate(gerentes['E-mail']):
    
    gerente = gerentes.loc[i , 'Gerente']
    relatorio = gerentes.loc[i , 'Relatório']
    
    mail = outlook.CreateItem(0)   
    mail.To = loop
    mail.Subject = 'relatorio de {}'.format(relatorio)
    mail.Body = '''
    Prezado {}, 
    Segue em anexo o Relatório de {}, conforme solicitado.
    Qualquer dúvida estou à disposição.
    Att.,
    '''.format(gerente , relatorio)
    
    attachment = r'C:\Users\caio.barbosa\Desktop\E-mail\Enviar E-mails.xlsx'
        
    mail.Attachments.Add(attachment)
    mail.Send()
    

 
    

   
```
